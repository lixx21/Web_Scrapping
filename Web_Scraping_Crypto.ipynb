```python
from bs4 import BeautifulSoup
import yfinance as yf
import pandas as pd
import requests

url = "https://coinmarketcap.com/trending-cryptocurrencies/"

data = requests.get(url).text

soup = BeautifulSoup(data)

title = soup.title
print("title from web : ", title)

tag = soup.h1
print("h1 from web : ",tag)

paragrahp = soup.p
print("First paragraph from web : ",paragrahp)

sibling1 = tag.next_sibling
print("Sibling of tag h1 : ", sibling1)

crypto_data = pd.DataFrame(columns = ["#","Coin", "Price", "24H", "7D", "30D", "Market Cap", "Volume"])
for i in soup.find("tbody").find_all("tr"):
  col = i.find_all("td")
  num = col[1].text
  coin = col[2].text
  price = col[3].text
  last24 = col[4].text
  last7d = col[5].text
  last30d = col[6].text
  market = col[7].text
  volume = col[8].text

  crypto_data = crypto_data.append({"#":num, "Coin":coin, "Price":price, "24H":last24, "7D":last7d, "30D":last30d, "Market Cap":market, "Volume":volume}, ignore_index=True)

crypto_data.head(20)
```
